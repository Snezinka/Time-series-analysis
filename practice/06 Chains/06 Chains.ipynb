## **Часть 1.** Поиск цепочек временного ряда.

Импорт библиотек и модулей

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import stumpy
from sklearn.metrics import mean_squared_error
```

В первом задании вам предстоит выполнить поиск эволюционирующих шаблонов временного ряда. Выполните считывание набора данных [Rotation Matrix](datasets/Rotation%20Matrix.txt), найдите в нем самую длинную цепочку и выведите индексы начала звеньев этой цепочки.

```python
fig, ax = plt.subplots(figsize=(16, 2))
ts = np.loadtxt('datasets/Rotation Matrix.txt')
ax.plot(ts)
```

```python
window_size = 80
matrix_profile = stumpy.stump(ts, m=window_size)
left_mp_index = matrix_profile[:, 2]
right_mp_index = matrix_profile[:, 3]
chains = stumpy.allc(left_mp_index, right_mp_index)
```

```python
long_chain_index = sorted(chains, key=len)[0]
long_chain = []
for i in long_chain_index:
  long_chain.append(ts[i:i + window_size])
```

## **Часть 2. Визуализация цепочек**

```python
fig, ax = plt.subplots(figsize=(16, 2))
ax.plot(ts)
for i in range(len(long_chain)):
  ax.plot(range(long_chain_index[i], long_chain_index[i] + window_size), long_chain[i], color = 'r')
```
